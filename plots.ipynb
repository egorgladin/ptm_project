```python
import os
import pickle
import matplotlib.pyplot as plt
```

```python
results_10_01 = 'results/10_01/'
results = []
for filename in os.listdir(results_10_01):
    with open(f'{results_10_01}{filename}', 'rb') as handle:
        res = pickle.load(handle)
        results.append(res)
```

```python
results_back_only = [res for res in results\
                     if res['parameters']['tau_sparse'] is None and res['parameters']['tau_decor'] is None]
results_back_decor = [res for res in results\
                      if res['parameters']['tau_sparse'] is None and res['parameters']['tau_decor'] is not None]
results_back_sparse = [res for res in results\
                      if res['parameters']['tau_sparse'] is not None and res['parameters']['tau_decor'] is None]
results_back_decor_sparse = [res for res in results\
                      if res['parameters']['tau_sparse'] is not None and res['parameters']['tau_decor'] is not None]
```

```python
back_only = sorted(results_back_only, key=lambda x: x['f1'])[-1]
back_decor = sorted(results_back_decor, key=lambda x: x['f1'])[-1]
back_sparse = sorted(results_back_sparse, key=lambda x: x['f1'])[-1]
back_decor_sparse = sorted(results_back_decor_sparse, key=lambda x: x['f1'])[-1]
```

```python
plt.plot(range(1, 20), back_decor['perplexity'][1:], label="Cглаживание+Декорреляция")

plt.xlabel("Iteration number")
plt.ylabel("Perplexity")
plt.legend()
# plt.show()
plt.savefig("Perplexity.png")
```

```python
plt.plot(range(1, 20), back_decor['sparsity'][1:], label="Cглаживание+Декорреляция")
plt.plot(range(1, 20), back_decor_sparse['sparsity'][1:], label="Все регуляризаторы")

plt.xlabel("Iteration number")
plt.ylabel("Sparsity")
plt.legend()
# plt.show()
plt.savefig("Sparsity.png")
```
